```python
import pandas as pd

df = pd.read_csv("../data/raw/train.csv")

# df.head() #intial rows display
# df.shape #dimensions
# df.columns 
df.info()
```

```python

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

df.dtypes
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Postal Code'])
```

Postal Code column contained 11 missing values, and since they are not important for Time Serires Analysis, those value rows are removed.

```python
df.isnull().sum()
```

Data integrity Maintained.

```python
df.duplicated().sum()
df = df.drop_duplicates()
df.info()
```

No duplicates found.

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order YearMonth'] = df['Order Date'].dt.to_period('M')
df.info()
```

```python
df['Postal Code'] = df['Postal Code'].astype(str)
df.info()
```
